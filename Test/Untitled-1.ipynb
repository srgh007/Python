```python
import os, sys
from time import ctime, strftime, time
# import time
dirfiles = []
cntr = 0     
for root, dirs, files in os.walk("C:\\ApachTomcat"):
    sz = os.path.getsize(root)
    dm = os.path.getatime(root)
    dmt = ctime(dm)
    # print(dmt)
    cntr = root.count(os.sep)
    # cntr = str(root).count('\\')
    if (cntr>0):
        print ("[{}] {}\n[{}]".format(cntr,root,dmt))
        for file in files:
            sz = os.path.getsize(root+"\\{}".format(file))
            print ("-"*cntr+" {} [{}]".format(file,sz))
```
